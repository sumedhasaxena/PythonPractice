```python
input_x = np.arange(-20, 20, step = 0.01)
input_x = input_x.reshape(input_x.shape[0], 1)

output_y = np.cos(input_x)

print(input_x.shape)
print(output_y.shape)
```

```python
def plot_xy(x : np.ndarray, y : np.ndarray) -> None:
    
    plt.figure(figsize = (15, 7))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(['Train'])
    plt.show()
    
    
def plot_xy_predict(x : np.ndarray, y : np.ndarray, prediction : np.ndarray) -> None:
    
    plt.figure(figsize = (15, 7))
    plt.plot(x, y)
    plt.plot(x, prediction)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(['Actual', 'Prediction'])
    plt.show()
```

```python
plot_xy(input_x, output_y)
```
